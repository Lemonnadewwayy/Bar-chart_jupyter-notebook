```python
import numpy as np
import matplotlib.pyplot as plt
real_day1 = [5453, 7548, 6543]    #C
real_day2 = [1840, 4013, 3421]   #oi
real_day3 = [1080, 1673, 2342]    #huh?
movies = ['HAHA', 'oi', 'huh?']
x=np.arange(3)
width=0.2
plt.bar(x,real_day1,width=width,label=movies[0])
bar2_x=x+width
plt.bar(bar2_x,real_day2,width=width,label=movies[1])
bar3_x=bar2_x+width
plt.bar(bar3_x,real_day3,width=width,label=movies[2])
def addlabels(x,y):
    for i in range(len(x)):
       plt.text(x[i],y[i],y[i], ha='center',va='bottom') 

addlabels(x, real_day1)
addlabels(bar2_x, real_day2)
addlabels(bar3_x, real_day3)
plt.legend()
plt.ylabel("票房(万元)")
xlabels=["第一天","第二天","第三天"]
plt.xticks(bar2_x,xlabels)
```
